# Matplot and Seaborn library

Both Matplotlib and Seaborn are powerful Python libraries for data visualization. 

# What is Matplotlib?
Matplotlib is a low level graph plotting library in python that serves as a visualization utility.

Matplotlib was created by John D. Hunter.

Matplotlib is open source and we can use it freely.

Matplotlib is mostly written in python, a few segments are written in C, Objective-C and Javascript for Platform compatibility.

# Seaborn
Visualize Distributions With Seaborn.


Seaborn is a library that uses Matplotlib underneath to plot graphs. It will be used to visualize random distributions.

# 1. Line Plot
Matplotlib:

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Plot
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='sin(x)')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot')
plt.legend()
plt.show()
```

Seaborn

```python
import seaborn as sns
import pandas as pd

# Data
data = pd.DataFrame({'x': x, 'y': y})

# Plot
sns.lineplot(data=data, x='x', y='y')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot')
plt.show()
```

# 2. Scatter Plot
Matplotlib:

```python
# Data
x = np.random.rand(50)
y = np.random.rand(50)

# Plot
plt.figure(figsize=(10, 6))
plt.scatter(x, y, c='blue', alpha=0.5)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot')
plt.show()
```

Seaborn

```python
# Data
data = pd.DataFrame({'x': x, 'y': y})

# Plot
sns.scatterplot(data=data, x='x', y='y')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot')
plt.show()
```

# 3. Bar Plot
Matplotlib:

```python
# Data
categories = ['A', 'B', 'C', 'D']
values = [4, 7, 3, 8]

# Plot
plt.figure(figsize=(10, 6))
plt.bar(categories, values, color='skyblue')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Plot')
plt.show()
```

Seaborn

```python
# Data
data = pd.DataFrame({'categories': categories, 'values': values})

# Plot
sns.barplot(data=data, x='categories', y='values')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Plot')
plt.show()
```

# 4. Histogram
Matplotlib:

```python
# Data
data = np.random.randn(500)

# Plot
plt.figure(figsize=(10, 6))
plt.hist(data, bins=30, alpha=0.7, color='green')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.show()
```

Seaborn

```python
# Plot
sns.histplot(data, bins=30, kde=False, color='green')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.show()
```

# 5. Box Plot
Matplotlib:

```python
# Data
data = [np.random.rand(50) * 100, np.random.rand(50) * 50, np.random.rand(50) * 75]

# Plot
plt.figure(figsize=(10, 6))
plt.boxplot(data, vert=True, patch_artist=True)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Box Plot')
plt.show()
```

Seaborn

```python
# Data
df = pd.DataFrame(data).melt(var_name='Category', value_name='Value')

# Plot
sns.boxplot(data=df, x='Category', y='Value')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Box Plot')
plt.show()
```

# 6. Heatmap
Matplotlib:

```python
# Data
data = np.random.rand(10, 12)

# Plot
plt.figure(figsize=(10, 6))
plt.imshow(data, cmap='viridis', aspect='auto')
plt.colorbar()
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Heatmap')
plt.show()
```

Seaborn

```python
# Plot
sns.heatmap(data, cmap='viridis')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Heatmap')
plt.show()
```

# 7  Count Plot (Bar plot for categorical data)
Matplotlib:

```python
# Data
categories = ['A', 'B', 'C', 'D', 'A', 'B', 'C', 'D', 'A', 'A']
counts = {cat: categories.count(cat) for cat in set(categories)}

# Plot
plt.figure(figsize=(10, 6))
plt.bar(counts.keys(), counts.values(), color='skyblue')
plt.xlabel('Categories')
plt.ylabel('Counts')
plt.title('Count Plot')
plt.show()
```

Seaborn

```python
# Plot
sns.countplot(x=categories)
plt.xlabel('Categories')
plt.ylabel('Counts')
plt.title('Count Plot')
plt.show()
```

# 8. Regression Plot
Matplotlib:

```python
# Data
x = np.random.rand(100)
y = 2 * x + np.random.randn(100) * 0.1

# Plot
plt.figure(figsize=(10, 6))
plt.scatter(x, y, label='Data points')
plt.plot(np.unique(x), np.poly1d(np.polyfit(x, y, 1))(np.unique(x)), color='red', label='Linear fit')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Regression Plot')
plt.legend()
plt.show()
```

Seaborn

```python
# Data
data = pd.DataFrame({'x': x, 'y': y})

# Plot
plt.figure(figsize=(10, 6))
sns.regplot(data=data, x='x', y='y')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Regression Plot')
plt.show()
```

# 9. Area Plot
Matplotlib:

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Plot
plt.figure(figsize=(10, 6))
plt.fill_between(x, y1, alpha=0.5, label='sin(x)')
plt.fill_between(x, y2, alpha=0.5, label='cos(x)')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Area Plot')
plt.legend()
plt.show()
```

Seaborn

```python
import seaborn as sns
import pandas as pd

# Data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
data = pd.DataFrame({'x': x, 'sin': y1, 'cos': y2})

# Plot
plt.figure(figsize=(10, 6))
sns.lineplot(data=data, x='x', y='sin', label='sin(x)')
sns.lineplot(data=data, x='x', y='cos', label='cos(x)')
plt.fill_between(data['x'], data['sin'], alpha=0.5)
plt.fill_between(data['x'], data['cos'], alpha=0.5)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Area Plot')
plt.legend()
plt.show()
```

# 10. Pie Chart
Matplotlib:

```python
# Data
labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.1, 0, 0, 0)  # explode 1st slice

# Plot
plt.figure(figsize=(8, 8))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.title('Pie Chart')
plt.show()
```

Seaborn:
Seaborn does not support pie charts directly.
